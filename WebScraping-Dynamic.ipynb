```python
#Import the libraries
import requests
import bs4
```

```python
from bs4 import BeautifulSoup
```

```python
#let's chose our dynamic website
url = 'https://www.makaan.com/nellore-residential-property/buy-property-in-nellore-city'
```

```python
#connect to the website
data = requests.get(url)
data
```

```python
data.text
type(data.text)
```

```python
soup = BeautifulSoup(data.text,'html')
soup
type(soup)
```

```python
#let's extract property name
g = soup.find()
g
```

```python
soup.find('a',attrs={'class':'seller-name'}).text
```

```python
s=soup.find_all('div',attrs={'class':'title-line'})
```

```python
names = []
for i in s:
    names.append(i.text)
print(names)
```

```python
names = [i.text for i in s]
```

```python
d = soup.find_all('div',attrs={'data-type':'price-link'})
price = [i.text.strip() for i in d]
print(price)
```

```python
e = soup.find_all('td',attrs={'class':'val'})
price = [i.text.strip() for i in e]
print(price)
print(len(price))
```

```python
nbath = soup.find_all('ul',attrs={'class':'listing-details'})
bath_info=[]
for i in nbath:
    if soup.find('li',attrs={'class':'keypoint','title':'Bathroom'})is None:
        bath_info.append("Not mentioned")
    else:
        bath_info.append(i,find('li',attrs={'class':'keypoint','title':'Bathrooms'}).text)
print(bath_info)
print(len(bath_info))
```
